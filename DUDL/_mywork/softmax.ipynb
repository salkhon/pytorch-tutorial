```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# manual softmax
z = [1, 2, 3]

num = np.exp(z)
denom = np.sum(num)
sigma = num / denom
print(sigma)
print(np.sum(sigma))
```

```python
# a visualization
z = np.random.randint(low=-5, high=15, size=25)

num = np.exp(z)
denom = np.sum(num)
sigma = num / denom

plt.plot(z, sigma, "ko")
plt.xlabel("Original number (z)")
plt.ylabel("Softmax $\sigma$")
plt.title(f"$\sum\sigma$ = {np.sum(sigma)}")
# plt.yscale("log")  # a linear transformation in log space. 
plt.show()
```

```python
# softmax activation class (callable)
softfun = nn.Softmax(dim=0)

sigmaT = softfun(torch.Tensor(z))  # list needs to be converted to tensor (using the class Tensor directly)

print(sigmaT)
```
